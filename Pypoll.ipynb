```python
import csv

# Path to the election data CSV file
file_path = "election_data.csv"

# Initialize variables to store analysis results
total_votes = 0
candidates = {}
winner = ""

# Open the election data CSV file and read in the data
with open(file_path, "r") as election_data:
    reader = csv.reader(election_data)
    
    # Skip the header row
    next(reader)
    
    # Loop through each row of the data
    for row in reader:
        # Increment the total vote count
        total_votes += 1
        
        # If the candidate has already been seen, increment their vote count
        if row[2] in candidates:
            candidates[row[2]] += 1
        # If the candidate has not been seen before, add them to the dictionary with a vote count of 1
        else:
            candidates[row[2]] = 1
    
    # Determine the winner based on the candidate with the most votes
    max_votes = 0
    for candidate, votes in candidates.items():
        if votes > max_votes:
            max_votes = votes
            winner = candidate
    
# Print the analysis results in the desired format
print("Election Results")
print("-------------------------")
print(f"Total Votes: {total_votes}")
print("-------------------------")
for candidate, votes in candidates.items():
    percentage = round(votes / total_votes * 100, 3)
    print(f"{candidate}: {percentage}% ({votes})")
print("-------------------------")
print(f"Winner: {winner}")
print("-------------------------")
```
