```python
import pandas as pd
from google.colab import files

# Upload the file
uploaded = files.upload()

# Read the CSV file using pandas
import io
data = pd.read_csv(io.BytesIO(uploaded['budget_data.csv']))
```

```python
import csv

# Initialize variables
total_months = 0
total_profit_loss = 0
prev_profit_loss = 0
profit_loss_change_list = []
max_profit_increase = 0
max_profit_increase_date = ""
max_profit_decrease = 0
max_profit_decrease_date = ""
```

```python
# Read data from file
with open('budget_data.csv', 'r') as csvfile:
    csvreader = csv.reader(csvfile, delimiter=',')
    header = next(csvreader) # skip header row

    # Loop through rows
    for row in csvreader:
        # Increment month count
        total_months += 1

        # Calculate total profit/loss
        profit_loss = int(row[1])
        total_profit_loss += profit_loss

        # Calculate profit/loss change
        if total_months > 1:
            profit_loss_change = profit_loss - prev_profit_loss
            profit_loss_change_list.append(profit_loss_change)

            # Check for max profit increase and decrease
            if profit_loss_change > max_profit_increase:
                max_profit_increase = profit_loss_change
                max_profit_increase_date = row[0]
            elif profit_loss_change < max_profit_decrease:
                max_profit_decrease = profit_loss_change
                max_profit_decrease_date = row[0]

        prev_profit_loss = profit_loss
```

```python
# Calculate average profit/loss change
average_profit_loss_change = sum(profit_loss_change_list) / len(profit_loss_change_list)

# Output results
print("Financial Analysis")
print("------------------")
print(f"Total Months: {total_months}")
print(f"Total: ${total_profit_loss}")
print(f"Average Change: ${average_profit_loss_change:.2f}")
print(f"Greatest Increase in Profits: {max_profit_increase_date} (${max_profit_increase})")
print(f"Greatest Decrease in Profits: {max_profit_decrease_date} (${max_profit_decrease})")
```
